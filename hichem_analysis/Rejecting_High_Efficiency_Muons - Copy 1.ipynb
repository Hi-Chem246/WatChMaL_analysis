## Remove the High Efficiency Muons  

```python
three_class_softmax = [softmaxes[0]] # Softmaxes for the 3-class model
```

```python
three_class_labels = [labels[0]] # Labels for the 3-class model 
```

```python
# Collapses electron and gamma probabilities together 

collapsed_class_scores_list, collapsed_class_labels_list = multi_collapse_test_output(three_class_softmax, three_class_labels, label_dict, ignore_type='$e$')

collapsed_class_labels_list = [collapsed_class_labels - 1 for collapsed_class_labels in collapsed_class_labels_list]

collapsed_class_scores_list = [collapsed_class_scores[:, 1:] for collapsed_class_scores in collapsed_class_scores_list]
```

```python
# Computes the ROC curve for the e/gamma vs muon case 

fprs, tprs, thrs = multi_compute_roc(collapsed_class_scores_list, collapsed_class_labels_list, 
                                                         true_label=muon_softmax_index_dict["non-mu"], 
                                                         false_label=muon_softmax_index_dict["mu"],
                                                         normalize=True)
```

```python
three_class_tpr = tprs[0] # Loads the true-positive rate (aka. efficiency) for the e/gamma vs muon case
```

```python
three_class_thr = thrs[0] # Loads the thresholds for the e/gamma vs muon case
```

```python
efficiency = 0.98 # The efficiency we want 
tolerance = 0.00001 # How much we are willing to tolerate 

# This loop searches through the tpr to find the efficiency closest to the one we want

for i in range(len(three_class_tpr)): 
    if (three_class_tpr[i] >= (efficiency-tolerance)) and (three_class_tpr[i] <= efficiency+tolerance):
        print(three_class_thr[i],i)
```

```python
threshold_index = 22722 # Index corresponding to the efficiency we want 
```

```python
threshold = three_class_thr[threshold_index] # Finds the threshold we need in the threshold data 
```

```python
print(threshold)
```

```python
three_class_softmax = softmaxes[0] # Softmaxes for the 3-class model
```

```python
three_class_labels = labels[0] # Labels for the 3-class model 
```

```python
idxs = [] # List of indices 

# This loop finds the indices of the softmax data which have a e+gamma score below the threshold  

for j in range(len(three_class_softmax)): 
    if (three_class_softmax[j,0]+three_class_softmax[j,1]) <= threshold:
        idxs.append(j)
        
idxs = np.array(idxs)
```

```python
print(len(idxs))
```

```python
# This cell keeps track of the kind of events that are removed 

event_list = []

for w in idxs:
    event_list.append(three_class_labels[w])
    
event_list = np.array(event_list)

gamma_events = []

e_events = []

muon_events = []

for r in event_list:
    if r == 0:
        gamma_events.append(r)
    if r == 1:
        e_events.append(r)
    if r == 2:
        muon_events.append(r)

print("Removed gamma events =", len(gamma_events))
print("Removed electron events =", len(e_events))
print("Removed muon events =", len(muon_events))
```

```python
softmaxes = softmaxes + [new_softmax]
labels = labels + [new_labels]
titles = titles + ["3-Class - Rejected Muons with 98% efficiency threshold"]
linecolor = linecolor + [c[3]]
linestyle = linestyle + ['-']
```
